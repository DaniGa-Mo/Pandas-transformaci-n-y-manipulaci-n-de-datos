```python
# Link 2 base de datos
import pandas as pd
url2 = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'
datos2 = pd.read_json(url2)
datos2
```

```python
# Normalizacion
datos2_normalizados = pd.json_normalize(datos2['dados_locacao'])
datos2_normalizados
```

```python
# Colectar los valores de las columnas y verificar
columnas2 = list(datos2_normalizados.columns)
columnas2
```

```python
# Destrincar las listas con explode
datos2 = datos2_normalizados.explode(columnas2[1:])
```

```python
# Resetear los index de las líneas
datos2.reset_index(drop=True,inplace=True)
# Observar el DataFrame
datos2
```

```python
# Verificar los tipos de datos con info
datos.info()
```

```python
# La columna numérica es el 'valor_aluguel'
datos2['valor_aluguel']
```

```python
# Iniciar la transformación
# Import de la biblioteca numpy
import numpy as np
# Remover los textos presentes en la base
# Cambiar las comas separadoras del decimal por punto
datos2['valor_aluguel'] = datos2['valor_aluguel'].apply(lambda x: x.replace('$ ', '').replace(' reais', '').replace(',','.').strip())
# Cambiar los tipos de datos para float
datos2['valor_aluguel'] = datos2['valor_aluguel'].astype(np.float64)
# Verificar la transformación
datos2.info()
```

```python
# Opción de substitución - necesario verificar el resultado de la substitución
# Fue necesario adicionar la barra '\' para ser considerados los paréntesis como caracteres literales
datos2['apartamento'].str.replace('\(blocoAP\)', '')

```

```python
# Realizar la substitución de los datos en la columna textual
datos2['apartamento'] = datos2['apartamento'].str.replace('\(blocoAP\)', '')
```

```python
# Visualizar el resultado final
datos2
```

```python
# Transformar para el tipo datetime definiendo el formato de fecha como DD/MM/AAAA ('%d/%m/%Y')
datos2['datas_combinadas_pagamento'] = pd.to_datetime(datos2['datas_combinadas_pagamento'], format='%d/%m/%Y')
datos2['datas_de_pagamento'] = pd.to_datetime(datos2['datas_de_pagamento'], format='%d/%m/%Y')

```

```python
# Visualizar el resultado
datos

```

```python
# Para contribuir en la solución del contexto es posible calcular la diferencia de días
# entre la fecha combinada y la fecha de pagamento con dt.days
datos2['atraso'] = (datos2['datas_de_pagamento'] - datos2['datas_combinadas_pagamento']).dt.days

# Observar el nuevo DataFrame
datos2

```
